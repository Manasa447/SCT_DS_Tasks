```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('national.csv')
data = data.dropna(subset=['population'])
age_data = data.groupby('age')['population'].sum()

plt.bar(age_data.index, age_data.values, color='orange')
plt.title('Bar Chart - Population by Age')
plt.xlabel('Age')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

plt.hist(data['population'], bins=10, color='green', edgecolor='black')
plt.title('Histogram - Population Distribution')
plt.xlabel('Population')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
